```python
# Create a subset of calculating the average price of diesel for every state

avg_diesel_price = df_price.groupby('state')['diesel'].mean()
```

```python
avg_diesel_price.head()
```

```python
# Convert the subset into a dataframe

df_avg_diesel_price = pd.DataFrame(avg_diesel_price)
```

```python
df_avg_diesel_price
```

```python
# Create a subset of calculating the e10 average price for every state

avg_e10_price = df_price.groupby('state')['e10'].mean()
```

```python
avg_e10_price
```

```python
# Convert the subset into a dataframe

df_avg_e10_price = pd.DataFrame(avg_e10_price)
```

```python
df_avg_e10_price
```

```python
# Calculate the average e5 price for every state
# Convert the subset into a dataframe

avg_e5_price = df_price.groupby('state')['e5'].mean()
df_avg_e5_price = pd.DataFrame(avg_e5_price)
df_avg_e5_price
```

```python
# Merge all 3 created subsets into one dataframe

df_avg_price = pd.merge(df_avg_diesel_price, df_avg_e10_price, on='state')
df_avg_price = pd.merge(df_avg_price, df_avg_e5_price, on='state')
df_avg_price
```
